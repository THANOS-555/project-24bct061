```python
import pandas as pd
```

```python
df = pd.read_csv("KDDTrain.csv")
```

```python
print("HEAD:")
display(df.head())

print("\nTAIL (last 10 rows):")
display(df.tail(10))

print("\nINFO:")
print(df.info())

print("\nDESCRIBE:")
display(df.describe())

print("\nNULL VALUES PER COLUMN:")
print(df.isnull().sum())
```

```python
df_no_na = df.dropna()
df_filled = df.fillna(20)
```

```python
df.fillna(20, inplace=True)
```

```python
if 'duration' in df.columns:
    df['duration'].fillna(20, inplace=True)
```

```python
if 'duration' in df.columns:
    print("\nSample of 'duration' column:")
    print(df['duration'].head())
```

```python
print("\nCOLUMN NAMES:")
print(df.columns)
```

```python
target_col = None
for col in df.columns:
    if col.lower() in ['class', 'label']:
        target_col = col
        break
```

```python
if target_col is not None:
    print(f"\nTarget/Label column detected: {target_col}")
    print("\nRows where attack is present (not 'normal'):")
    attacks = df[df[target_col] != 'normal']
    display(attacks.head())

    print("\nGroup by class/label:")
    print(df.groupby(target_col).count())
else:
    print("\n No 'class' or 'label' column detected. Please check df.columns and adjust.")
```

```python
from sklearn.preprocessing import LabelEncoder

cat_cols = df.select_dtypes(include=['object']).columns
print("\nCategorical columns to encode:", list(cat_cols))

for col in cat_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])

print("\nAfter Label Encoding, info:")
print(df.info())
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
print("\nNumeric columns to scale:", list(num_cols))
```

```python
print("\nSample after scaling:")
display(df.head())
```

```python
output_name = "KDDTrain_processed.csv"
df.to_csv(output_name, index=False)
```
